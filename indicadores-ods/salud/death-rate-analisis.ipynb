```python
#check that entidad y periodo are on all data frames

master_df=dfs[0]

for df in dfs[1:]:
    common_columns=[]
    for element in list(master_df):
        if element in list(df): common_columns.append(element)
    master_df=pd.merge(master_df, df, on=common_columns)
```
